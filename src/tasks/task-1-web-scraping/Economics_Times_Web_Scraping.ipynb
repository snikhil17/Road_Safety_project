```python
import bs4
import requests
import warnings
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from bs4 import BeautifulSoup

warnings.filterwarnings('ignore')
```

```python
url = "https://economictimes.indiatimes.com/"
```

```python
response = requests.get(url = url)
```

```python
keywords = ['traffic', 'accident', 'incidents', 'suicide', 'murders', 
            'attempt to murder', 'car accident', 'attack']
```

```python
def get_data(url, **kwargs):
    response = requests.get(url = url)
    if response.status_code == 200:
        news_data = list()
        images_data = list()
        videos_data = list()
        result = {}
        try:
            if kwargs['news']:
                print("Getting News Data :: ")
                for page_no in range(1,21):
                    news_url = url + "/news/" + str(page_no)
                    news_response = requests.get(url = news_url)
                    
                    content = news_response.content
                    soup = BeautifulSoup(content, "html.parser")
                    contentD = soup.find_all("div", class_ = "contentD")
                    for i in range(len(contentD)):
                        print("Page No. : {} ---- News Article : {}".format(page_no,i))
                        article_link = contentD[i].find('a')['href']
                        article_url = url + article_link
                        article_response = requests.get(url = article_url)
                        article_content = article_response.content
                        article_soup = BeautifulSoup(article_content, 'html.parser')
            
                        try:
                            summary = article_soup.find_all('div', class_ = "topPart clearfix tac fixedOnLoad")[0].find('h1').text
                        except:
                            summary = None

                        try:
                            datetime = article_soup.find_all('div', class_ = 'dtc vam artByline')[0].find('time').text
                        except:
                            datetime = None

                        try:
                            image = article_soup.find_all('div', class_ = "imgBox")[0].find('figure').find('img')['src']
                        except:
                            image = None

                        try:
                            full_article = article_soup.find_all('article', class_ = 'artData clr')[0].find_all('div', class_ = "artText")
                        except:
                            full_article = None
                        news_data.append([summary, datetime, image, full_article])
                        
                if result.get('news_data') is None:
                    result['news_data'] = news_data
                else:
                    pass
                        
        except:
            pass
        
        
        try:
            if kwargs['images']:
                print("Getting Images Data :: ")
                for page_no in range(1,5):
                    images_url = url + "/photos/" + str(page_no)
                    images_response = requests.get(url = images_url)
                    content = images_response.content
                    soup = BeautifulSoup(content, "html.parser")
                    contentD = soup.find_all("div", class_ = "contentD")
                    if len(contentD) > 0: 
                        for i in range(len(contentD)):
                            print("Page No. : {} ---- News Article : {}".format(page_no,i))
                            image_link = contentD[i].find('a')['href']
                            image_url = url + image_link
                            image_response = requests.get(url = image_url)
                            image_content = image_response.content
                            image_soup = BeautifulSoup(image_content, 'html.parser')
                            try:
                                image = image_soup.find_all('img', class_ = 'slideImg')[0]['src']
                            except:
                                pass
                            images_data.append((i,image, image_url))
                    else:
                        pass
                if result.get('image_data') is None:
                    result['image_data'] = images_data
                else:
                    pass
        except:
            pass
        try:
            if kwargs['videos']:
                print("Getting Videos Data :: ")
                for page_no in range(1,21):
                    videos_url = url + "/videos/" + str(page_no)
                    videos_response = requests.get(url = videos_url)
                    content = videos_response.content
                    soup = BeautifulSoup(content, "html.parser")
                    contentD = soup.find_all("div", class_ = "contentD")

                    if (len(contentD))>1:
                        for i in range(len(contentD)):
                            print("Page No. : {} ---- News Article : {}".format(page_no,i))
                            video_link = contentD[i].find('a')['href']
                            video_url = url + video_link
                            id_ = video_url[-12:-4]
                            video_link = "https://economictimes.indiatimes.com/news/videodash.cms?autostart=1&msid="+id_
                            videos_data.append((i, id_, video_url, video_link))
                    else:
                        pass
                    
                if result.get('videos_data') is None:
                    result['video_data'] = videos_data
                else:
                    pass
        except:
            pass
        
        
    return result
```

```python
%%time
data = {}
for key in keywords:
    try:
        if len(key)>1:
            key = key.replace(" ", "-")
        get_url = url + "topic/" + key
        response = requests.get(url = get_url)
        if (response.status_code == 200):
            print("Getting Data of Keyword :: {}".format(key))
            if data.get(key) is None:
                data[key] = get_data(get_url, news = True, images = True, videos = True)
            else:
                pass
        else:
            continue
    except:
        raise KeyError("The keyword not found!")
```
